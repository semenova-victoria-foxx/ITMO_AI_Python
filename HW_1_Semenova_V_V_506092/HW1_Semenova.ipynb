#ЧАСТЬ 1. Работа со строками и исключениями (4 балла)

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None
```

## Демонстрация работы №1

```python
examples = ["12", "-7 ", "12a", ""]
for example in examples:
    result = parse_int(example)
    print(f'"{example}" -> {result}')
```

# ЧАСТЬ 2. Циклы и работа со списками (6 баллов)

```python
def count_positive(nums: list[int]) -> int:
    count = 0
    for num in nums:
        if num > 0:
            count += 1
    return count
def sum_even(nums: list[int]) -> int:
    total = 0
    for num in nums:
        if num % 2 == 0:
            total += num
    return total
def first_divisible(nums: list[int], k: int) -> int | None:
    for num in nums:
        if num % k == 0:
            return num
    return None
```

## Демонстрация работы №2

```python
import random
# Создаём список из 25 случайных чисел от -10 до 20
numbers = [random.randint(-10, 20) for _ in range(25)]
print("Список чисел:", numbers)
print("Количество положительных:", count_positive(numbers))
print("Сумма чётных:", sum_even(numbers))
k = 3
result = first_divisible(numbers, k)
print(f"Первое число, делящееся на {k}:", result)
```

#ЧАСТЬ 3. Лямбда и сортировки (2 балла)

```python
items = [
    {"название": "А", "цена": 120, "рейтинг": 4.7},
    {"название": "В", "цена": 90, "рейтинг": 4.2},
    {"название": "С", "цена": 150, "рейтинг": 4.9}
]
```

## Демонстрация работы №3

### Сортировка по цене (по возрастанию)

```python
by_price = sorted(items, key=lambda x: x["цена"])
print("По цене (возрастание):", by_price)
```

### Сортировка по рейтингу (по убыванию)

```python
by_rating = sorted(items, key=lambda x: x["рейтинг"], reverse=True)
print("По рейтингу (убывание):", by_rating)
```

#ЧАСТЬ 4. Мини-ООП (2 балла)

```python
class Шоколад:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Цена должна быть положительной!")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        """Возвращает цену со скидкой в процентах"""
        return self.price * (100 - percent) / 100
```

## Демонстрация работы №4

```python
try:
    # Создаём объекты
    шок1 = Шоколад("Алёнка", 120)
    шок2 = Шоколад("Milka", 150)

    print(f"{шок1.name}: {шок1.price} руб.")
    print(f"Со скидкой 15%: {шок1.discount(15)} руб.\n")

    print(f"{шок2.name}: {шок2.price} руб.")
    print(f"Со скидкой 20%: {шок2.discount(20)} руб.")


except ValueError as e:
    print("Ошибка:", e)
```
